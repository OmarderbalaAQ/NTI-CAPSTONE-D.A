# 3-data cleaning 

### mixed units (unit price)

```python
df["Unit price"]= df["Unit price"].str.replace(' USD', '', regex=False)
```

```python
df["Unit price"].unique()
```

### incorrect data type (unit price)

```python
df[["Unit price"]]=df[["Unit price"]].astype('float64')
```

```python
df["Unit price"].dtype
```

### inconsistent time format

```python
# function to convert 12hh-mm to 24hh:mm
def convert_to_24_hour(time_str):
    if ' PM' in time_str or ' pm' in time_str :
        hour, minute = time_str.split(' PM')[0].split('- ')
        hour = int(hour) + 12
        return f"{hour}:{minute}"
    elif ' AM' in time_str or ' am' in time_str:
        hour, minute = time_str.split(' AM')[0].split('-')
        return f"{hour}:{minute}"
    else:
        return time_str  # Assuming already in 24-hour format

```

```python
df.head()
```

```python
df["Time"]=df["Time"].apply(convert_to_24_hour)
df["Time"].head()
```
